## 1.1 Why Study Databases?

Every organization depends on data. Colleges store student records. Hospitals track patients and appointments. Retail companies manage customers, products, and orders. Manufacturers track inventory, suppliers, machines, and production schedules.

A **database** is a structured collection of related data. Databases help organizations store, organize, search, update, protect, and share data.

Databases matter because they allow people to answer important questions, such as:

- Which students are enrolled in a course?
- Which customers placed orders last month?
- Which products are low in inventory?
- Which employees worked on a project?
- Which machines need maintenance?

In this book, we will focus mostly on **relational databases** and **SQL**, which is the language commonly used to work with relational databases.

## 1.2 Data vs. Information

Although people often use the words interchangeably, **data** and **information** are not exactly the same.

**Data** are raw facts or values.

Examples:

| StudentID | FirstName | LastName | Major |
|---|---|---|---|
| 1001 | Maya | Johnson | IST |
| 1002 | Carlos | Rivera | Business |
| 1003 | Amina | Patel | Engineering |

**Information** is data that has been organized or interpreted so that it is useful.

Examples:

- There are 3 students in the sample.
- One student is majoring in Business.
- Student 1003 is Amina Patel.

A database helps turn raw data into useful information.

## 1.3 What Is a Database?

A **database** is an organized collection of related data.

The word **organized** is important. A random list of names is not usually enough. A useful database has structure, rules, and relationships.

For example, a college registration database might include:

- Students
- Courses
- Instructors
- Course sections
- Enrollments

These are related. A student enrolls in a course section. A course section is taught by an instructor. A course belongs to a department.

A database allows these related pieces of data to be stored together in a structured way.

## 1.4 What Is a DBMS?

A **Database Management System**, or **DBMS**, is software used to create, manage, and interact with databases.

Examples of database management systems include:

- SQLite
- MySQL
- PostgreSQL
- Microsoft SQL Server
- Oracle Database

The database is the stored data.  
The DBMS is the software that manages the database.

A DBMS helps users:

- Add new data
- Update existing data
- Delete data
- Search data
- Protect data
- Control who can access data
- Prevent inconsistent or invalid data

## 1.5 Databases vs. Spreadsheets

Spreadsheets are useful tools, especially for small tasks. However, as data becomes larger, more connected, or shared by many users, spreadsheets can become difficult to manage.

| Spreadsheet | Database |
|---|---|
| Good for small, simple datasets | Good for larger, structured datasets |
| Often edited manually | Often managed using rules and queries |
| Easy to duplicate inconsistent data | Designed to reduce unnecessary duplication |
| Limited control over relationships | Designed to model relationships |
| Not ideal for many users editing at once | Better for shared organizational systems |

For example, a spreadsheet might store customer and order data in one large table. A database would usually separate customers, orders, and products into different but related tables.

## 1.6 Tables, Rows, and Columns

Relational databases organize data into **tables**.

A table is made of:

- **Columns**: categories or fields of data
- **Rows**: individual records
- **Cells**: individual values

Example table: `Student`

| StudentID | FirstName | LastName | Major |
|---|---|---|---|
| 1001 | Maya | Johnson | IST |
| 1002 | Carlos | Rivera | Business |
| 1003 | Amina | Patel | Engineering |

In database terminology:

- A table is often called a **relation**.
- A row is often called a **tuple**.
- A column is often called an **attribute**.

We will use both everyday language and formal database language throughout this book.

## 1.7 A First Look at SQL

**SQL** stands for **Structured Query Language**.

SQL is used to communicate with relational databases. SQL can be used to:

- Create tables
- Insert data
- Retrieve data
- Update data
- Delete data
- Control access to data

A simple SQL query looks like this:

```sql
SELECT *
FROM Student;
```

This means: show all columns and rows from the `Student` table.

## 1.8 Why Structure Matters

Imagine storing all college data in one spreadsheet:

| StudentName | Major | CourseName | Instructor | Room |
|---|---|---|---|---|
| Maya Johnson | IST | Intro to Databases | Dr. Lee | 202 |
| Carlos Rivera | Business | Intro to Databases | Dr. Lee | 202 |
| Amina Patel | Engineering | Web Development | Dr. Gomez | 118 |

This might work at first, but problems appear quickly:

- The instructor's name may be repeated many times.
- A room change might require updates in many rows.
- A student's name could be typed differently in different places.
- It becomes difficult to track which students, courses, instructors, and rooms are truly separate things.

A database helps solve these problems by organizing data into related tables.

## 1.9 Common Types of Databases

Although this book focuses on relational databases, it is useful to know that there are several types of databases.

| Type | Basic Idea | Example Use |
|---|---|---|
| Relational database | Data stored in related tables | Student records, sales systems |
| Document database | Data stored as flexible documents | Web applications, product catalogs |
| Graph database | Data stored as nodes and relationships | Social networks, recommendation systems |
| Key-value database | Data stored as key-value pairs | Caching, fast lookups |
| Object-oriented database | Data stored as objects | Specialized software systems |

Relational databases remain widely used because they are powerful, structured, and supported by SQL.

## 1.10 Key Terms

| Term | Meaning |
|---|---|
| Data | Raw facts or values |
| Information | Data organized into a meaningful form |
| Database | Organized collection of related data |
| DBMS | Software used to manage a database |
| Table | Structure that stores data in rows and columns |
| Row | A single record in a table |
| Column | A field or attribute in a table |
| SQL | Language used to interact with relational databases |
| Query | A request for data from a database |
| Primary key | A column or set of columns that uniquely identifies a row |
